```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Salary_Data_cleansed.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
def title_simplifier(title):
    if 'data scientist' in title.lower():
        return 'data scientist'
    elif 'data engineer' in title.lower():
        return 'data engineer'
    elif 'analyst' in title.lower():
        return 'analyst'
    elif 'machine learning' in title.lower():
        return 'machine learning'
    elif 'manager' in title.lower():
        return 'manager'
    elif 'director' in title.lower():
        return 'director'
    else:
        return 'na'

def senioriy(title):
    if 'sr' in title.lower() or 'senior' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
        return 'senior'
    elif 'jr' in title.lower() or 'junior' in title.lower() or 'jr.' in title.lower():
        return 'jr'
    else:
        return 'na'
    
```

```python
df['job_simple'] = df['Job Title'].apply(title_simplifier)
```

```python
df['job_simple'].value_counts()
```

```python
df['seniority'] = df['Job Title'].apply(senioriy)
df['seniority'].value_counts()
```

```python
#fix los angeles state
df['job_state'] =df['job_state'].apply(lambda x: x.strip() if x.strip().lower() != 'los angeles' else 'CA')
df['job_state'].value_counts()
```

```python
df['desc_len'] = df['Job Description'].apply(lambda x: len(x))
df['desc_len']
```

```python
df['num_comp'] = df['Competitors'].apply(lambda x: len(x.split(',')) if x!='-1' else 0)
```

```python
df['Competitors']
```

```python
#hourly wage to annual. //multiply 2000(annual working hours) to hourly wage and divide by 1000(since we nee the first 2 digits)
#df['min_salary1'] = df.apply(lambda x: x.min_salary*2 if x.hourly==1 else x.min_salary,axis=1)
df['min_salary'] = df.apply(lambda x: x.min_salary*2 if x.hourly==1 else x.min_salary,axis=1)
df['max_salary'] = df.apply(lambda x: x.max_salary *2 if x.hourly==1 else x.max_salary,axis=1)
```
